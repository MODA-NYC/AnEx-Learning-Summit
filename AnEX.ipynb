## Dealing with missing values

In the real-world data rarely do not require cleaning. Missing data can occur during data extraction or collection. Missing values must be handled carefully because the technique(s) used may reduce the quality of model performance. Thus, the approach to dealing with missing data is largely dependent on the nature of our data and the missing values. 

Python machine learning estimators (SKLearn) expect numeric values and that all have a meaning.

#### Some approaches for dealing with missing data

- Drop NULL or missing values
- Fill Missing Values
- Predict Missing values with an ML Algorithm

### Fill missing values

This is a common approach of handling missing values. Missing values of a feature are replaced with a test statistic like mean, median or mode. Letâ€™s suppose we have a missing value of age in the boston data set. Then the below code will fill the missing value with the 30.

```python
df['Lot Frontage'].unique()
```

```python
df['Lot Frontage'] = df['Lot Frontage'].fillna(df['Lot Frontage'].mean())
```

```python
df['Lot Frontage'].unique()
```

### Create dummy variables

- Dummy variables are a set of binary (0 or 1) variables that each represent a single class from a categorical feature.

```python
#Select all object data types in the dataset

df_objs = df.select_dtypes(include='object')
```

```python
df_objs.info()
```

```python
df_num = df.select_dtypes(exclude='object')
```

```python
df_num.info()
```

#### Covert to dummy variables

```python
df_objs = pd.get_dummies(df_objs,drop_first=True)
```

```python
df_objs.head()
```

```python
#Concatenate the objects and numbers dataframes back together

concat_df = pd.concat([df_num,df_objs],axis=1)
```
